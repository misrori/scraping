```python
def get_n_facts_of_today(n=10):
    # Get the current date
    facts =[]
    counter = 0
    current_date = datetime.datetime.now().strftime("%m/%d")
    link = f"http://numbersapi.com/{current_date}/date"
    while len(facts)!=n:
        new_fact = requests.get(link).text
        if new_fact not in facts:
            facts.append(new_fact)
        else:
            counter +=1

        if counter == n*2:
          break
    return(facts)
```

```python
get_n_facts_of_today(n=30)
```
